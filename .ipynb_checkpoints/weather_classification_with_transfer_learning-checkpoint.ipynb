```python
#Prepare data
weather_path = "./images"
global_image_size = (150,150)
global_labels = os.listdir(weather_path)

def missed_data_report(missed_data):
    missed_data_count = len(missed_data)
    if missed_data_count > 0:
        print(str(missed_data_count) + " data has lost!")
        print(missed_data)
    else:
        print("No data loss!")

def load_images_from_file(path,path_labels):
    missed_data = []
    data = []
    labels = []
    for label_no,label in enumerate(path_labels,start = 0):
        label_path = os.path.join(path,label)
        for image_name in os.listdir(label_path):
            try:
                image_path = os.path.join(label_path,image_name)
                image = cv2.imread(image_path)
                image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
                image_from_array = Image.fromarray(image,'RGB')
                resized_image = image_from_array.resize(global_image_size)
                data.append(np.array(resized_image))
                labels.append(label_no)
            except Exception as e:
                missed_data.append(image_name)
    missed_data_report(missed_data)
    return data,labels

data,labels = load_images_from_file(weather_path,global_labels)
print(global_labels)
```

```python
weathers = np.array(data)
labels = np.array(labels)
```

```python
base_model.trainable = False
```

```python
inputs = keras.Input(shape=(global_image_size[0],global_image_size[1],3))
x = base_model(inputs,training=False)
x = keras.layers.MaxPooling2D(pool_size=2)(x)
x = keras.layers.Conv2D(filters=64,kernel_size=2,padding="same",activation="relu")(x)
x = keras.layers.MaxPooling2D(pool_size=2)(x)
x = keras.layers.Dropout(0.2)(x)
x = keras.layers.Flatten()(x)
x = keras.layers.Dense(500,activation="relu")(x)
x = keras.layers.Dropout(0.2)(x)
outputs = keras.layers.Dense(num_classes,activation="softmax")(x)
model = keras.Model(inputs,outputs)
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
callback = keras.callbacks.EarlyStopping(monitor="loss", patience=3)
```

```python
history = model.fit(X_train,y_train,batch_size=32, epochs=3, verbose = 1,callbacks=[callback], validation_data=(X_test,y_test))
```

```python
from sklearn.metrics import roc_curve
y_pred = model.predict(X_test,batch_size=32)
y_pred = np.argmax(y_pred,axis=1)
y_test_cm = np.argmax(y_test,axis=1)
confusion_matrix(y_test_cm,y_pred)
```

```python
def get_conf_mat_str(cm):
    my_str = "\n\nKarƒ±≈üƒ±klƒ±k Matrisi\n------------------\n"
    for i in range(cm.shape[0]):
        my_str += "[ "
        for j in range(cm.shape[1]):
            if j != cm.shape[1] - 1:
                my_str += str(cm[i,j]) + ", "
            else:
                my_str += str(cm[i,j]) + " "
        my_str += "]\n"
    return my_str
conf_mat_str = get_conf_mat_str(conf_mat)
print(conf_mat_str)
```

```python
def get_dl_history_str(history):
    my_str = "\n\nEpoch ba≈üƒ±na Accuracy - Loss oranlarƒ±\n--------------------------------------\n"
    count = len(history["accuracy"])
    for i in range(count):
        my_str += "Adƒ±m " + str(i + 1) + "/" + str(count) + " -> Accuracy : " + str(history["accuracy"][i]) + ", Loss : " + str(history["loss"][i])+"\n"  
    return my_str

his_str = get_dl_history_str(history.history)
print(his_str)
```
